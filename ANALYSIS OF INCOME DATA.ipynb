```python
# REPLACE SALARY VALUE "<=50K" AND ">50K" WITH 0.0 AND 1.0 RESPECTIVELY .
income["salary"].unique()
```

```python
income["salary"].value_counts()
```

```python
# FIRST METHOD 
def salary_data(sal):
    
    if sal == "<=50K":
        return 0
    else:
        return 1
    
income["encoded-salary"] = income["salary"].apply(salary_data) 
income.head()
```

```python
# SECOND METHOD
income.replace ( to_replace=["<=50K",">50K"] , value=[0,1] , inplace=True )
income.head()
```
